### <a id=5>train_cv</a>

```python
trained_model, result_dict = model_train.train_cv(X=train_std_data['X_train_std_1'], y=train_data['target_train_1'],
                                                  model=Lasso(alpha=1000),
                                                  cv_type='loo',
                                                  metric_best='RMSE_val',
                                                  data_name='orig')

pd.DataFrame([result_dict])
```

```python
import joblib

joblib.dump(value=trained_model,
            filename="models/std/lasso_std_1.pkl")
```

```python
trained_model, result_dict = model_train.train_cv(X=train_data['X_train_1'], y=train_data['target_train_1'],
                                                  model=HuberRegressor(alpha=100, epsilon=5, max_iter=1000),
                                                  cv_type='stratify',
                                                  metric_best='RE_val',
                                                  n_splits=8,
                                                  data_name='orig')

pd.DataFrame([result_dict])
```

```python
trained_model, result_dict = model_train.train_cv(X=train_data['X_train_1'], y=train_data['target_train_1'],
                                                  model=LinearRegression(),
                                                  cv_type='kf',
                                                  metric_best='RE_val',
                                                  n_splits=5,
                                                  data_name='orig')

pd.DataFrame([result_dict])
```

```python
trained_model, result_dict = model_train.train_cv(X=train_data['X_train_1'], y=train_data['target_train_1'],
                                                  model=LinearRegression(),
                                                  cv_type='ts',
                                                  metric_best='R2_val',
                                                  train_size=5,
                                                  val_size=5)

pd.DataFrame([result_dict])
```
